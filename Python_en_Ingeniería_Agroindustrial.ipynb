## ¿Qué es Python?

En términos sencillos, Python es un lenguaje de programación **de alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Es famoso por ser:

* **Legible:** No necesitas ser un genio de la informática para entender qué hace una línea de código.
* **Versátil:** Sirve para analizar datos, controlar sensores, automatizar máquinas o incluso crear inteligencia artificial.
* **Gratuito y Abierto:** Tiene una comunidad gigantesca que crea "librerías" (herramientas listas para usar) que facilitan el trabajo técnico.

## Ejemplos de Código (Sencillos y Útiles)

Aquí tienes tres scripts básicos que podrías usar en el día a día agroindustrial:

### 1. Balance de Masa Simple

Imagina que quieres saber cuánto jugo concentrado obtendrás tras un proceso de evaporación.

```
# Cálculo de balance de masa en evaporación
# Entrada = Salida + Evaporado

jugo_inicial = 500  # kg
brix_inicial = 10   # % de azúcar
brix_final = 50     # % de azúcar deseado

# Fórmula: Masa1 * Conc1 = Masa2 * Conc2
jugo_final = (jugo_inicial * brix_inicial) / brix_final
agua_evaporada = jugo_inicial - jugo_final

print(f"Resultado: Obtendrás {jugo_final} kg de concentrado.")
print(f"Se deben evaporar {agua_evaporada} kg de agua.")
```

### 2. Clasificador de Madurez (Lógica booleana)

Un pequeño sistema que decide si un lote de materia prima es apto para proceso basado en grados Brix y firmeza.

```
grados_brix = 14.5
firmeza = 6.2  # Escala de 1 a 10

if grados_brix >= 12 and firmeza >= 5:
    estado = "Apto para exportación"
elif grados_brix >= 12 and firmeza < 5:
    estado = "Apto para procesamiento local (mermeladas)"
else:
    estado = "Rechazado / Maduración insuficiente"

print(f"Decisión de calidad: {estado}")
```

### 3. Estimación de Carga Térmica

Un cálculo rápido para saber cuánta energía necesitas para enfriar un producto en un silo o tanque.
Para esto usamos la fórmula de calor sensible: $Q = m \cdot C_p \cdot \Delta T$

```
# Datos del producto (ejemplo: leche)
masa = 1000          # kg
cp_leche = 3.89      # kJ/kg°C
temp_inicial = 25    # °C
temp_final = 4       # °C

# Cálculo de energía (Q)
delta_t = temp_inicial - temp_final
energia_kj = masa * cp_leche * delta_t

print(f"Calor a retirar: {energia_kj} kJ para alcanzar los 4°C.")
```
